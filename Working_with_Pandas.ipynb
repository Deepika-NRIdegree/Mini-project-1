```python
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Price'] = pd.to_numeric(df['Price'])
df['Discount'] = pd.to_numeric(df['Discount'])

df['Discount'] = df['Discount'].fillna(0)
```

```python
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
df['NetPrice']
```

```python
  df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
df['TotalSale']
```

```python
filtered_df = df[df['TotalSale'] >= 20]
```

```python
df
```

```python
filtered_df['PaymentType'] = filtered_df['PaymentType'].astype('category')
```

```python
df['PaymentType']
```

```python
grouped_df = filtered_df.groupby('ProductCategory').agg({
    'Quantity': 'sum',
    'TotalSale': 'sum'
})
```

```python
sorted_grouped_df = grouped_df.sort_values(by='TotalSale', ascending=False)
```

```python
top_3_categories = sorted_grouped_df.head(3)
print(top_3_categories)
```

```python
# productcatagory
multi_grouped_df = filtered_df.groupby(['StoreID', 'ProductCategory']).agg({
    'TotalSale': 'mean'
})
```

```python
highest_average_TotalSale_for_each_store = multi_grouped_df.reset_index().sort_values(['StoreID', 'TotalSale'], ascending=[True, False]).drop_duplicates('StoreID', keep='first')
```

```python
df['StoreID']
```

```python
summary_df = highest_average_TotalSale_for_each_store.rename(columns={'TotalSale': 'AvgTotalSale'}).reset_index(drop=True)
print(summary_df)
```
